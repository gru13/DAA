```python
from copy import deepcopy

def printMatrixLayer(mat):
    re = deepcopy(mat[0])
    for a in mat[1:]:
        for b in range(3):
            re[b].append('|')
            re[b].extend(a[b])
    print("------------------------------------------------------------")
    for a in re:
        print(' '.join(a))
    print("------------------------------------------------------------")

def score(Input, Output):
    return sum(Input[i][j] != Output[i][j] for i in range(3) for j in range(3) if Output[i][j] != ' ')

def Branch_Bound(Input, Output):
    frontier = [(Input, 0)]  # (node, cost)
    explored = set()
    while frontier:
        node, cost = frontier.pop(0)
        explored.add(tuple(map(tuple, node)))  # Convert 2D list to tuple for hashing
        if node == Output:
            return node
        x, y = next((i, j) for i in range(3) for j in range(3) if node[i][j] == ' ')
        for dx, dy in ((1, 0), (-1, 0), (0, 1), (0, -1)):
            nx, ny = x + dx, y + dy
            if 0 <= nx < 3 and 0 <= ny < 3:
                new_node = deepcopy(node)
                new_node[x][y], new_node[nx][ny] = new_node[nx][ny], new_node[x][y]
                new_cost = cost + 1
                new_node_tuple = tuple(map(tuple, new_node))
                if new_node_tuple not in explored:
                    frontier.append((new_node, new_cost))
        frontier.sort(key=lambda x: score(x[0], Output) + x[1])  # Sort by (heuristic + cost)
    return None  # No solution found

Output = [
    ['1', '2', '3'],
    ['4', '5', '6'],
    ['7', '8', ' ']
]

Input = [
    ['1', '4', '5'],
    ['2', ' ', '7'],
    ['6', '3', '8'],
]

res = Branch_Bound(Input, Output)
if res:
    printMatrixLayer([res])
else:
    print("No solution found.")
```
